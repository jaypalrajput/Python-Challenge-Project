Q3> 
Pearson Corelation Coefficient

```python
import numpy as np #import numpy as np to perform math calculations and expressions
import statistics as stat #import statistics as stat for statistics numeric measures

def person_correlation(x,y):  #Function to evaluate Pearson corelation coefficient
    if len(x)!=len(y):     #To compare both variables has values else return false
        return 0.0
    n=len(x)          #storing length of x in n
    xmean=stat.mean(x)  # storing mean of x into xmean
    ymean=stat.mean(y)   # storing mean of y into ymean
    
    for i in range(n):    
        numrate=(x[i]-xmean)*(y[i]-ymean) #subtracting xmean from x[i] and ymean from y[i] and multipy equation to calculate numerator
        denomitor=(np.sqrt(x[i]-xmean)**2)*(np.sqrt(y[i]-ymean)**2) #Multiplying square root of each xmean and ymean with exponentiation
    r=numrate/denomitor  #Deviding numerate by denominator
    return r          #Returning r
    
x=[1,2,3,4,5,6,7,8,9]
y=[1,2,3,2,3,4,3,4,5]
  
print(person_correlation(x,y))  #Calling Function

x=[1,2,3,4,5,6,7,8,9]
y=[1,2,3,2,3,4,3]
  
print(person_correlation(x,y)) #Calling Function
```

#Q.5 Write a progam to accept a string s from user that contains (,),{,} and [,] in it.

```python
def matched(str):  #The function that check bracket balanced and unbalanced
    count = 0
    
    for i in str:
        if i == "(":  #Compare index opening of bracket if true count +1  
            count += 1
        elif i == ")": #Compare index of closing bracket if true count minus and assigned by -1 
            count -= 1
        elif i == "{":   
            count += 1
        elif i == "}":
            count -= 1
        elif i == "[":
            count += 1
        elif i == "]":
            count -= 1
        if count < 0:   #If count = 0 means bracket not closed properly
            return False
    return count == 0 
matched("{jjjjj(jhdhddh)[ghhhhhh]}")
```
